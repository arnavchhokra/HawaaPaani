Weighted Arithmetic Index Method

```python
si = [20,7.25,7.5,1500,4,10,5000,10]
```

```python
def wqi(df):
    wqi = []
    for i in df.index:
        idx = 0
        qiwi = 0
        w = 0
        for c in columns:
            if pd.isna(df[c][i]) or df[c][i] == '':
                qi = 0
            else:
                qi = (float(df[c][i])/si[idx])
            wi = 1/si[idx]
            qiwi += (qi*wi)
            w += wi
            idx += 1
        wqi.append(qiwi/wi)
    return wqi
```

```python
WQI = wqi(df)
```

```python
df['WQI'] = WQI
```
